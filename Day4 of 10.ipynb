```python
df_mean= df.copy()
```

```python
df_1=df
```

```python
df_2=df
```

```python
df_mean['RHOB']=df_mean['RHOB'].fillna(df_mean['RHOB'].mean())
df_mean['GR']=df_mean['GR'].fillna(df_mean['GR'].mean())
df_mean['PHIE']=df_mean['PHIE'].fillna(df_mean['PHIE'].mean())
display('after filling nans',df_mean)
```

```python
df_liner= df_1.copy()
```

```python
df_liner['GR']=df_liner['GR'].interpolate(method ='linear', limit_direction ='backward')
df_liner['RHOB']=df_liner['RHOB'].interpolate(method ='linear', limit_direction ='backward')
df_liner['PHIE']=df_liner['PHIE'].interpolate(method ='linear', limit_direction ='forward')
display('after filling nans',df_liner)
```

```python
df_polynomial= df_2.copy()
```

```python
df_polynomial['GR']=df_liner['GR'].interpolate(method ='polynomial',  order=2)
df_polynomial['RHOB']=df_liner['RHOB'].interpolate(method ='polynomial',  order=2)
df_polynomial['PHIE']=df_liner['PHIE'].interpolate(method ='polynomial',  order=2)
display('after filling nans',df_polynomial)
```
